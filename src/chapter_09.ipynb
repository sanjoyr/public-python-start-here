# Chapter 9: Writing Reusable Logic

## Problem 1: Why Is Copy-Paste Logic a Bad Idea?

```
population = 91276115
if population > 50000000:
    print("Highly populated")
else:
    print("Less populated")
```

## Problem 2: How Do We Package Logic into a Function?

```
def population_category(population):
    if population > 50000000:
        return "Highly populated"
    else:
        return "Less populated"
```

## Problem 3: How Do We Reuse the Function?

```
print(population_category(91276115))
print(population_category(112374333))
```

## Problem 4: What Are Positional and Keyword Arguments?

```
def population_report(state, population):
    return state + " has population " + str(population)
```

```
population_report("West Bengal", 91276115)
population_report(state="Maharashtra", population=112374333)
```

## Problem 5: How Do Functions Handle Flexible Inputs?

```
def average_population(*populations):
    total = 0
    for p in populations:
        total += p
    return total / len(populations)
```

```
average_population(91276115, 112374333, 104099452)
```

## Problem 6: How Do We Handle Optional Named Data?

```
def population_summary(**states):
    for state, population in states.items():
        print(state, ":", population)
```

```
population_summary(
    West_Bengal=91276115,
    Maharashtra=112374333
)
```

## Problem 7: Why Functions Matter in Data Science

```
def calculate_literacy_rate(literate_pop, total_pop):
    """Calculates the percentage of literacy for a given population."""
    return (literate_pop / total_pop) * 100
```
