# Chapter 4: Working with Numbers

## Problem 1: What Information Do We Already Have?

```
total_population = 1210854977
number_of_states = 35
```

## Problem 2: How Do We Calculate an Average?

```
average_population = total_population / number_of_states
print(average_population)
```

## Problem 3: Why Does the Result Have Decimals?

```
print(type(average_population))
```

## Problem 4: What Other Calculations Can We Do?

```
population_increase = 5000000
new_population_estimate = total_population + population_increase
print(new_population_estimate)
```

```
double_population = total_population * 2
print(double_population)
```

```
half_population = total_population / 2
print(half_population)
```

## Problem 5: How Do Expressions Work in Python?

```
result = (total_population / number_of_states) / 1000000
print(result)
```

```
result = total_population / number_of_states / 1000000
print(result)
```

## Problem 6: Why Order of Operations Matters

```
wrong_result = total_population / (number_of_states / 1000000)
print(wrong_result)
```

## Problem 7: Using Calculations to Create New Knowledge

```
average_population_millions = average_population / 1000000
print(average_population_millions)
# Professional output formatting
print(f"Average: {average_population_millions:.2f} Million")
```
