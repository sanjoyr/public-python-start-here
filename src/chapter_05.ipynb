## Problem 3: How Does Python Make a Decision?

```
if state_population > 50000000:
    print("This is a highly populated state")
```

## Problem 4: What If the Condition Is Not True?

```
if state_population > 50000000:
    print("This is a highly populated state")
else:
    print("This is not a highly populated state")
```

## Problem 5: Can We Check Multiple Conditions?

```
if state_population > 100000000:
    print("Extremely high population")
elif state_population > 50000000:
    print("High population")
else:
    print("Moderate population")
```

## Problem 6: Why Order Matters in Multiple Conditions

```
if state_population > 50000000:
    print("High population")
elif state_population > 100000000:
    print("Extremely high population")
```

## Problem 7: Avoiding Too Many `if` Statements

```
if state_population > 100000000:
    print("Extremely high")

if state_population > 50000000:
    print("High")
```

## Problem 8: Using Boolean Logic to Clarify Decisions

```
is_highly_populated = state_population > 50000000
print(is_highly_populated)
```
