## Problem 2: How Do We Repeat the Same Logic for Each Value?

## Problem 3: How Do We Apply a Decision Inside a Loop?

```
for population in state_populations:
    if population > 50_000_000:
        print("Highly populated:", population)
    else:
        print("Moderate population:", population)
```

## Problem 4: Why Is Indentation So Important?

## Problem 5: When Do We Stop or Skip During a Loop?

```
for population in state_populations:
    if population < 70_000_000:
        continue
    print("Large state:", population)
```

```
for population in state_populations:
    if population > 110_000_000:
        print("Very large state found")
        break
```

## Problem 6: What About `while` Loops?

```
count = 0

while count < len(state_populations):
    print(state_populations[count])
    count = count + 1
```
