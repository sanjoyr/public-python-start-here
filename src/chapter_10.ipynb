```
try:
    population = int("not_a_number")
    result = population / 2
except ValueError:
    print("Population data is invalid")
```

## Problem 6: Combining Validation with Error Handling

```
def safe_half_population(value):
    if value is None or value == "":
        return "Missing data"

    try:
        value = int(value)
    except ValueError:
        return "Invalid population"

    if value <= 0:
        return "Invalid population"

    return value / 2
```

```
safe_half_population("91276115")
safe_half_population(None)
safe_half_population("")
safe_half_population("abc")
```
