## Problem 2: How Do We Pair a Name with a Value?

```
state_population = {
    "West Bengal": 91276115,
    "Maharashtra": 112374333
}
```

## Problem 3: How Do We Look Up Data by Name?

```
population_wb = state_population["West Bengal"]
print(population_wb)
```

## Problem 4: What Happens If a Key Does Not Exist?

```
# This will raise an error
state_population["Kerala"]
```

```
if "Kerala" in state_population:
    print(state_population["Kerala"])
else:
    print("State not found")
```

## Problem 5: How Do We Process All States in a Dictionary?

```
for state, population in state_population.items():
    print(state, "has population", population)
```

## Problem 6: Why Dictionaries Matter for Data Science

```
# A list of dictionaries = A table of data
census_table = [
    {"state": "West Bengal", "population": 91276115},
    {"state": "Bihar", "population": 104099452}
]
print(census_table)
```
