## Problem 1: How Do We Normally Transform Data?

```
populations = [91276115, 104099452, 112374333, 72147030]

large_states = []
for p in populations:
    if p > 90000000:
        large_states.append(p)

large_states
```

## Problem 2: Can We Describe the Transformation Directly?

```
large_states = [p for p in populations if p > 90000000]
large_states
```
